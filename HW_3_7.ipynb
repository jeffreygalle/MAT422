## 3.7.1 Mathematical Formulation

In simple neural network the inputs x1 and x2 are combined with weights (w1, w2) and a bias b to predict an output (y^).

The objective here is to adjust teh weights and biases to minimize the difference between predicted outputs and actual data.

Neural networks iteratively adjust the weights based on error, gradually learning to match the outputs with the target data over training iterations.

Formulation representation: layer computation is expressed in matrix form:
z^(l) = W^(l)a^(l-1) + b^(l)  where W^(l) represents weights, b^(l) is thr bias for layer l.




## 3.7.2 Activation Function

Activation functions determine a node's output based on the input it receives for tasks such as classification. In biological terms, activation function is similar to the decision of whether or not a neuron fires.



1.   Step Function


*   Has as a binary output (0 or 1), where the output switches on if the input exceeds a certain threshold value.
*   Useful for classification tasks


2.   ReLU (Rectified Linear Unit):


*   ReLU outputs the input directly if it is positive; otherwise, it outputs 0.
*   Useful in deep neural networks due to its ability for efficient training on complex datasets.

3. Sigmoid Function


*   Maps inputs to a value *between* 0 and 1.
*   Given by Ïƒ(x) = 1/(1+e^-x)

4. Softmax Function


*   Converts an input vector of K values into a probability distribution, where each output value is proportional to the input's scale
*   Commonly applied in  output layer of neural network, particularly for multiple class classification tasks, it transforms scores into interpretable probabilities.










```python
import numpy as np
import matplotlib.pyplot as plt

# simple example functions
def step_function(x):
    return np.where(x >= 0, 1, 0)

def relu(x):
    return np.maximum(0, x)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def softmax(x):
    e_x = np.exp(x - np.max(x))  # For numerical stability
    return e_x / e_x.sum(axis=0)

# input data create
x = np.linspace(-10, 10, 100)

step_values = step_function(x)
relu_values = relu(x)
sigmoid_values = sigmoid(x)

# softmax with multiple-input
inArr = [1, 2, 3, 4, 5]
x_multi = np.array(inArr)
softmax_values = softmax(x_multi)

plt.figure(figsize=(6, 4))

plt.subplot(3, 1, 1)
plt.plot(x, step_values, label='Step Function')
plt.title('Step function')
plt.grid(True)


plt.subplot(3, 1, 2)
plt.plot(x, relu_values, label='ReLU Function')
plt.title('Rectified Linear Units function')
plt.grid(True)

plt.subplot(3, 1, 3)
plt.plot(x, sigmoid_values, label='Sigmoid Function')
plt.title('Sigmoid functio')
plt.grid(True)
plt.tight_layout()
plt.show()
```

## 3.7.3 Cost Function


*   Cost function measures the error between predicted outputs (^y_n) and actual outputs (y_n)
*   For regression tasks, the cost function commonly used is the **least squares method**. Least squares calculates the average squared difference between the predicted and actual values.
*   For binary classification tasks, the cross entropy cost function is used. Cross entropy (also known as log loss) penalizes incorrect predictions more heavily by calculating the negative log of the predicted probability


## 3.7.4 Backpropagation

Method used for training neural networks by adjusting weights based on the error from each training iteration.

Goal: The goal is to minimize the cost function J with respect to network parameters (weights W and biases b)


*   Derivatives of J w.r.t. each weight and bias are calculated using gradient descent.

## 3.7.5 Backpropagation Algorithm



1.   The weights and biases are initialized (often randomly)
2.   The input vector x is fed through the network to compute the activations and output ^y.
3. Gradient Descent updates: Use the gradients calculated through backpropagation and update weights and biases.
